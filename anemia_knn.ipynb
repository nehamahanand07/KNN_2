```python
#https://www.kaggle.com/datasets/ehababoelnaga/anemia-types-classification

import pandas as pd
import numpy as np
df=pd.read_csv('anemia.csv')
df
```

```python
#missing value checking
df.isna().sum()
```

```python
#x and y separating as arrays
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1].values
y
```

```python
x.ndim
```

```python
y.ndim
```

```python
#converting training and testing data
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_test
```

```python
x_train
```

```python
y_train
```

```python
y_test
```

```python
#normalizing
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#algorithm implementation
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier(n_neighbors=7)
knn_model.fit(x_train,y_train)
y_pred=knn_model.predict(x_test)
y_pred

```

```python
y_test
```

```python
print(knn_model.predict(scaler.transform([[5.6,30.1,1.8,4,4,3.9,4.1,11,45,89,24,145,12.5,0.26]])))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#displaying confusion matrix
#labels order check from y_train

from sklearn.metrics import ConfusionMatrixDisplay
labels=['Normocytic normochromic anemia','Healthy','Other microcytic anemia','Normocytic hypochromic anemia','Macrocytic anemia','Iron deficiency anemia','Thrombocytopenia',
        'Leukemia','Leukemia with thrombocytopenia']
cmd=ConfusionMatrixDisplay(cm,display_labels=labels)
cmd.plot()
```

```python
#accuracy score
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
score
```

```python
#STEP : CLASSIFICATION REPORT based on training and testing data , not entire data
from sklearn.metrics import classification_report
cr=classification_report(y_test,y_pred)
print(cr)
```
